```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("matches.csv")
df.head()
df.info()
df.describe()
# Matches per season
season_counts = df['season'].value_counts()
season_counts.plot(kind='bar', figsize=(8,4))
plt.title("Matches per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
# Top teams by wins
top_teams = df['winner'].value_counts()
top_teams.plot(kind='bar', figsize=(8,4))
plt.title("Top Teams by Wins")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.show()
# Toss decision (bat/field)
df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Toss Decision (Bat vs Field)")
plt.show()
# Toss winner vs match winner probability
toss_match = (df['toss_winner'] == df['winner']).sum()
total = len(df)
percentage = (toss_match / total) * 100
percentage
```
